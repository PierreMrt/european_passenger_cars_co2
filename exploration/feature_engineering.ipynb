```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
INPUT_DATA = '../data/data_processed.csv'
```

```python
df = pd.read_csv(INPUT_DATA)
```

#### Distribution des colonnes numériques 

* pour voir comment gérer les valeurs manquantes

* voir si il y a besoin de normalisation

```python
# Liste des colonnes numériques
num_cols = df.select_dtypes(include=['number']).columns

# Affiche un histogramme pour chaque colonne numérique
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.show()
```

#### Pour gérer les valeurs manquantes:
* les valeurs manquantes pour l'âge seront remplacées par la médiane car la distribution est constante

* Pour le cylindre (ec), les valeurs manquantes seront remplacées par le mode car la distribution est multimodale et qu'utiliser la moyenne ou la médianne pourrait modifier la structure des données

* On remplacera les valeurs manquantes de la colonne 'm (kg)' (masse du véhicule) par la médiane, car la distribution de cette colonne est unimodale mais asymétrique vers la droite (présence de véhicules plus lourds).

#### Comment uniformiser les variables ?

* ec (cm3) : Distribution multimodale. Il convient d'utiliser une standardisation (StandardScaler), ou mise à l’échelle min-max

* ep (KW) : Multimodale. Idem

* m (kg) : Asymétrique à droite. StandardScaler car peu d'outliers

* Fuel consumption : Multimodal. RobustScaler plutôt que StandardScaler du fait de la présence d'outliers

* age_months : Répartition relativement uniforme. MinMaxScaler ou StandardScaler.
